```python
from datetime import datetime

def get_day_of_week(date_str):
    date = datetime.strptime(date_str, '%Y-%m-%d')
    return date.strftime('%A')
```

```python
data_dict = {}
years, months, days, day_of_week = [[] for _ in range(4)]
release = []
for i in jsonnn['items']:
    album_name = i['name']
    album_id = i['id']

    if album_name not in data_dict:
        data_dict[album_name] = album_id

        release_date = i['release_date']
        release.append(release_date)
        print(release_date)
        day_date = get_day_of_week(release_date)
        release_date_split = i['release_date'].split('-')
        year, month, day = release_date_split[0], release_date_split[1], release_date_split[2]
        years.append(year), months.append(month), days.append(day), day_of_week.append(day_date)
```

```python
def get_month_names(months):
    month_dict = {
        '01': 'January',
        '02': 'February',
        '03': 'March',
        '04': 'April',
        '05': 'May',
        '06': 'June',
        '07': 'July',
        '08': 'August',
        '09': 'September',
        '10': 'October',
        '11': 'November',
        '12': 'December'
    }
    month_names = [month_dict[month] for month in months]
    return month_names

months_names = get_month_names(months)
```

```python
months_names
```

```python
album_ids = []
for key in data_dict:
    value = data_dict[key]
    album_ids.append(value)
```

```python
def get_track_duration(ms):
    mins, secs = divmod(ms / 1000, 60)

    track_length = f'{int(mins)}:{int(secs)}'
    return track_length
```

```python
album_name = list(data_dict.keys())
```

```python
album_name
```

```python
a_name, release_d, release_y, release_m, release_m_str, release_days, release_days_str, a_id, songs, track_id, \
number_of_album_songs, track_length, track_minutes, track_seconds, track_nums, number_artists, artist, dance_lvl, \
energy_lvl, keys, loudness_lvl, speechiness_lvl, acousticness_lvl,  instrumentalness_lvl, liveness_lvl, \
valence_lvl, tempo_lvl = [[] for _ in range(27)]

idx = 0
for k in data_dict:
    id = data_dict[k]
    print('Album: ', k, '| ID: ', id)

    #get album tracks
    r = requests.get(f'https://api.spotify.com/v1/albums/{id}/tracks?limit=50', headers=headers)
    print(r.status_code)
    print('===========')
    album_tracks = r.json()
    num_tracks = len(album_tracks['items'])

    artist_idx = 0
    #loop through album songs
    for i in album_tracks['items']:
        
        song_id = i['id']
        song_name = i['name']

        #get song details
        song_data = requests.get(f'https://api.spotify.com/v1/audio-features/{song_id}', headers=headers)

        audio_features = song_data.json()

        danceability, energy, key, loudness, speechiness, acousticness,  instrumentalness, \
        liveness, valence, tempo = audio_features['danceability'], audio_features['energy'], \
        audio_features['key'], audio_features['loudness'], audio_features['speechiness'], \
        audio_features['acousticness'], audio_features['instrumentalness'], audio_features['liveness'], \
        audio_features['valence'], audio_features['tempo']
 
        #song length data
        track_duration = get_track_duration(i['duration_ms'])
        track_duration_split = track_duration.split(':')
        minutes = track_duration_split[0]
        secs = track_duration_split[1]
    

        #number of artists on song
        num_artists = len(i['artists'])

        #song track number on album
        track_num = i['track_number']

        #building up artist(s) feature
        if num_artists > 1:
            names = '' + i['artists'][0]['name']
            for singer in i['artists'][1:]:
                names += ',' + singer['name']
            artist.append(names)
        else:
            artist.append(i['artists'][0]['name'])
        

        a_name.append(k), release_d.append(release[idx]), release_y.append(years[idx]), release_m.append(months[idx]), \
        release_m_str.append(months_names[idx]), release_days.append(days[idx]), release_days_str.append(day_of_week[idx]), \
        a_id.append(id), songs.append(song_name), track_id.append(song_id), number_of_album_songs.append(num_tracks), \
        track_length.append(track_duration), track_minutes.append(minutes), track_seconds.append(secs), track_nums.append(track_num), \
        number_artists.append(num_artists), dance_lvl.append(danceability), energy_lvl.append(energy), keys.append(key), \
        loudness_lvl.append(loudness), speechiness_lvl.append(speechiness), acousticness_lvl.append(acousticness), \
        instrumentalness_lvl.append(instrumentalness), liveness_lvl.append(liveness), valence_lvl.append(valence), tempo_lvl.append(tempo)
    
    idx += 1
    time.sleep(2)
    print('ALBUM DONE')
```

```python

# create a dictionary to store column names and lists
df_data = {
    "Album": a_name,
    "Release Date": release_d,
    "Release Year": release_y,
    "Release Month": release_m,
    "Release Month (Name)": release_m_str,
    "Release Day": release_days,
    "Release Day (Name)": release_days_str,
    "Album ID": a_id,
    "Number of Album Songs": number_of_album_songs,
    "Song Name": songs,
    "Artist": artist,
    "Song ID": track_id,
    "Track Duration": track_length,
    "Track Duration (Minutes)": track_minutes,
    "Track Duration (Seconds)": track_seconds,
    "Track Number": track_nums,
    "Number of Artists": number_artists,
    "Danceability": dance_lvl,
    "Energy": energy_lvl,
    "Key": keys,
    "Loudness": loudness_lvl,
    "Speechiness": speechiness_lvl,
    "Acousticness": acousticness_lvl,
    "Instrumentalness": instrumentalness_lvl,
    "Liveness": liveness_lvl,
    "Valence": valence_lvl,
    "Tempo": tempo_lvl
}

# create a DataFrame with the given data
df = pd.DataFrame(df_data)
```

```python
# Split 'Artist' column by comma and expand into separate columns
df[['Artist 1', 'Artist 2', 'Artist 3']] = df['Artist'].str.split(',', expand=True)

# Replace empty strings with NaNs
df = df.replace('', np.nan)

# Fill NaNs in 'Artist 2' and 'Artist 3' columns with the string 'N/A'
df[['Artist 2', 'Artist 3']] = df[['Artist 2', 'Artist 3']].fillna(value='N/A')

# Replace any 'N/A' values in 'Artist 3' column with NaNs if 'Artist 2' column is also 'N/A'
df['Artist 3'] = df.apply(lambda row: np.nan if (row['Artist 2'] == 'N/A' and row['Artist 3'] == 'N/A') else row['Artist 3'], axis=1)
```
